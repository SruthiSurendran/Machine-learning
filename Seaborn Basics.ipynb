# IRIS Dataset

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.shape
```

```python
pd.unique(df['species'])
```

```python
sns.distplot(df['sepal_length'])
```

```python
sns.distplot(df['sepal_length'],hist=False)
```

```python
sns.distplot(df['sepal_length'],kde=False)
```

```python
sns.countplot(df['species'],palette='summer')
```

```python
##scatter
sns.scatterplot(df['sepal_length'],df['petal_length'])
```

```python
##pair plot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='species',diag_kind='kde',palette='husl')
```

```python
sns.pairplot(df,hue='species',diag_kind='hist',palette='summer')##if 
```

```python
sns.boxplot(df['sepal_length'])
```

```python
sns.boxplot(data=df,x='species',y='sepal_length',palette='winter')##palette is color ie:winter,summer ...
```

```python
sns.swarmplot(data=df,x='species',y='sepal_length',palette='winter')##in virginica one dot is so far ie is the outer layer
```

```python
sns.violinplot(data=df,x='species',y='sepal_length',palette='winter')
```

```python
pd.isnull(df)
```

```python
sns.heatmap(pd.isnull(df))
```

```python
sns.heatmap(df.corr())
```
